```python
arr = list(map(int, input().split()))
ops = 0
for i in range(1, len(arr)):
    if arr[i] <= arr[i-1]:
        need = arr[i-1] + 1
        ops += need - arr[i]
        arr[i] = need
print(ops)
```

```python
weights = list(map(int, input().split()))
limit = int(input())
weights.sort()
l, r = 0, len(weights)-1
boats = 0
while l <= r:
    if weights[l] + weights[r] <= limit:
        l += 1
    r -= 1
    boats += 1
print(boats)
```

```python
n = int(input())
nums = list(map(int, input().split()))
if n == 1:
    print(0)
else:
    jumps = curr_end = farthest = 0
    for i in range(n-1):
        farthest = max(farthest, i + nums[i])
        if i == curr_end:
            jumps += 1
            curr_end = farthest
            if curr_end >= n-1:
                break
    print(jumps)
```
