```python
a = list(map(int, input().split()))
n = len(a)
result = [-1] * n
stack = []
for i in range(n):
    while stack and a[stack[-1]] >= a[i]:
        stack.pop()
    if stack:
        result[i] = stack[-1]
    stack.append(i)
print(result)
```

```python
from collections import deque

n = int(input())
left, right = deque(), deque()

def balance():
    if len(left) < len(right):
        left.append(right.popleft())
    elif len(left) > len(right) + 1:
        right.appendleft(left.pop())

for _ in range(n):
    cmd = input().split()
    if cmd[0] == '+':
        right.append(cmd[1])
        balance()
    elif cmd[0] == '*':
        left.append(cmd[1])
        balance()
    elif cmd[0] == '-':
        if not left:
            while right:
                left.append(right.popleft())
        print(left.popleft())
        balance()
```
